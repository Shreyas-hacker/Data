```python
Stocks = google.groupby('Date')['Open'].sum() #Makes Date the index and sums everything tgt
Stocks1 = google.groupby('Date')['High'].sum()
Stocks2 = google.groupby('Date')['Low'].sum()
Stocks3 = google.groupby('Date')['Close'].sum()
Stocks4 = google.groupby('Date')['Adj Close'].sum()
Stocks5 = google.groupby('Date')['Volume'].sum()
```

```python
Stocks.index = pd.to_datetime(Stocks.index)
year = Stocks.index.to_period('Y')
year =year.to_timestamp() # makes the dates into year timestamps
```

```python
Data = pd.DataFrame(Stocks)
Data1 = pd.DataFrame(Stocks1)
Data2 = pd.DataFrame(Stocks2)
Data3 = pd.DataFrame(Stocks3)
Data4 = pd.DataFrame(Stocks4)
Data5 = pd.DataFrame(Stocks5)
```

```python
A=Data.set_index(year)
B=Data1.set_index(year)
C=Data2.set_index(year)
D=Data3.set_index(year)
E=Data4.set_index(year)
F=Data5.set_index(year)
```

```python
Open = A.groupby('Date')['Open'].sum()
High = B.groupby('Date')['High'].sum()
Low = C.groupby('Date')['Low'].sum()
Close = D.groupby('Date')['Close'].sum()
AdjClose = E.groupby('Date')['Adj Close'].sum()
Volume = F.groupby('Date')['Volume'].sum()
```

```python
Open = pd.DataFrame(Open)
High = pd.DataFrame(High)
Low = pd.DataFrame(Low)
Close = pd.DataFrame(Close)
AdjClose = pd.DataFrame(AdjClose)
Volume = pd.DataFrame(Volume)
```

```python
plt.figure(figsize = (25,35))
plt.suptitle('Google Historical Data from 2004 to 2020')

plt.subplot(311)
plt.plot(Open,'bo')
plt.title('Google Open')
plt.xlabel('Date')
plt.ylabel('Open')
plt.grid(True, color = 'k')

plt.subplot(312)
plt.plot(High,'bo')
plt.title('Google High')
plt.xlabel('Date')
plt.ylabel('High')
plt.grid(True, color = 'k')

plt.subplot(313)
plt.plot(Low,'bo')
plt.title('Google Low')
plt.xlabel('Date')
plt.ylabel('Low')
plt.grid(True, color = 'k')
```

```python
plt.figure(figsize = (25,35))
plt.suptitle('Google Historical Data from 2004 to 2020')

plt.subplot(311)
plt.plot(Close,'bo')
plt.title('Google Close')
plt.xlabel('Date')
plt.ylabel('Close')
plt.grid(True, color = 'k')

plt.subplot(312)
plt.plot(AdjClose,'bo')
plt.title('Google Adj Close')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.grid(True, color = 'k')

plt.subplot(313)
plt.plot(Volume,'bo')
plt.title('Google Volume')
plt.xlabel('Date')
plt.ylabel('Volume in 100000')
plt.grid(True, color = 'k')
```
